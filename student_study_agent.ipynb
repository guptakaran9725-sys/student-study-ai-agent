```
# Student Study Assistant AI Agent
# Kaggle Day 5 Capstone Project

import numpy as np

student_data = {
    "name": "Karan",
    "study_hours": 2,
    "subjects": {
        "Math": 65,
        "Science": 72,
        "English": 85,
        "Computer": 55
    }
}

def find_weak_subjects(subject_scores):
    weak = {sub: score for sub, score in subject_scores.items() if score < 70}
    return weak

def make_study_plan(hours, weak_subjects):
    plan = {}
    if not weak_subjects:
        return {"message": "Great! No weak subjects. Keep revising all topics!"}

    time_per_subject = hours / len(weak_subjects)

    for subject in weak_subjects:
        plan[subject] = f"Study {round(time_per_subject, 1)} hours"

    return plan

def ai_suggestions(weak_subjects):
    if not weak_subjects:
        return "Aap kaafi accha perform kar rahe ho! Bas regular revision karo."

    tips = "Weak Subjects Improve Tips:\n"
    for subject in weak_subjects:
        tips += f"ðŸ‘‰ {subject}: Previous year questions + daily practice 30 min\n"

    return tips

print("ðŸ“˜ Student Study Assistant AI Agent\n")

print("Student Name:", student_data["name"])
print("\nSubject Scores:", student_data["subjects"])

weak = find_weak_subjects(student_data["subjects"])
print("\nâ— Weak Subjects:", weak if weak else "None")

plan = make_study_plan(student_data["study_hours"], weak)

print("\nðŸ“ Personalized Study Plan:")
for sub, rec in plan.items():
    print(f"- {sub}: {rec}")

print("\nðŸ’¡ AI Suggestions:")
print(ai_suggestions(weak))
```
